```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
# z=f(x,y) 를 이용해 등고선 플롯 그리기

def f(x,y):
    return np.sin(x)**10 + np.cos(10+y*x)*np.cos(x)
```

등고선 플롯은 plt.contour 함수로 만들 수 있는데...
이 함수는 x값의 격자, y값의 격자, z값의 격자의 세개의 인수를 취한다
x와y는 플롯의 위치를 나타내고 z는 등고선의 높이를 나타낸다

1차원 배열로 부터 2차원 그리드를 만드는 np.meshgrid 함수를 이용하자

```python
x = np.linspace(0,5,50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x,y)
Z = f(X, Y)
```

```python
plt.contour(X,Y,Z, colors='black'); # 음수 값은 점선으로 양수값은 실선으로 표시된다
```

```python
# 20개의 동일한 간격의선을 더 그리고 cmap으로 색상표를 지정하여 색을 입히자

plt.contour(X,Y,Z, 20, cmap='RdGy'); # RdGy = Red-Gray의 약어
```

```python
plt.cm.<tab> # 가능 색상 목록 확인하기
```

```python
# 색으로 채워진등고선 그리기

plt.contourf(X,Y,Z, 20, cmap='RdGy')
plt.colorbar();
```

plt.imshow() 함수를 이용하면 이미지로 등고선 표현 가능

주의사항
(1) x와 y그리드를 받지 않으므로 플롯에 이미지의 extent=[xmin, xmax, ymin, ymax]를 직접 입력
(2) origin을 lower로 변경해야 한다
(3) plt.axis(aspect='image')를 통하여 x와 y의 단위를 일치시켜야 한다

```python
plt.imshow(Z, extent=[0,5,0,5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image');
```

```python
# 등고선 자체에 레이블 붙이기

contours = plt.contour(X,Y,Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0,5,0,5], origin='lower', cmap='RdGy', alpha=0.5)
plt.colorbar();
```
