```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
# plt.axes로 축 삽입하기

ax1 = plt.axes()
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])
```

```python
# fig.add-axes()로 두개의 세로로 배치된 축 만들기

fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],
                  xticklabels=[], ylim=(-1.2, 1.2))

ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],
                  ylim=(-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x))
```

```python
# plt.subplot 로 간단한 서브플롯 그리드 만들기

for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5,0.5,str((2,3,i)),
            fontsize=18, ha='center')
```

```python
# fig.add_subplot을 이용하여 만들기

fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4) # 간격 조정하기
for i in range(1,7):
    ax = fig.add_subplot(2,3,i)
    ax.text(0.5,0.5,str((2,3,i)),
            fontsize=18, ha='center')
```

```python
#plt.subplots 로 한번에 전체 그리드 만들기

fig, ax = plt.subplots(2,3, sharex='col', sharey='row')
```

```python
# 축이 2차원 배열 안에 있어 [행, 열]로 인덱싱 할 수 있음

for i in range(2):
    for j in range(3):
        ax[i, j].text(0.5, 0.5, str((i, j)),
                          fontsize=18, ha='center')
        
fig
```

```python
#plt.GridSpec 을 활용하여 복잡한 배치 만들기

grid = plt.GridSpec(2,3, wspace=0.4, hspace=0.4)
grid
```

```python
plt.subplot(grid[0,0])
plt.subplot(grid[0,1:])
plt.subplot(grid[1,:2])
plt.subplot(grid[1,2]);
```

```python
# 정규분포 데이터 만들기

mean = [0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean,cov,3000).T

# GridSpec으로 축 설정하기

fig = plt.figure(figsize=(6,6))
grid = plt.GridSpec(4,4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1,0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1,1:], yticklabels=[], sharex=main_ax)

fig
```

```python
# 메인 축에 점 산포하기

main_ax.plot(x,y, 'ok', markersize=3, alpha=0.2)
```

```python
# 보조 축상에 히스코그램 만들기

x_hist.hist(x, 40, histtype='stepfilled',
           orientation='vertical', color='gray')
x_hist.invert_yaxis()
y_hist.hist(y, 40, histtype='stepfilled',
           orientation='horizontal', color='gray')
y_hist.invert_xaxis()
fig
```
