1.7.2 성과 측정: 훈련데이터와 테스트 데이터

이 데이터로 머신러닝 모델을 만들고 새로운 데이터의 품종을 예측하려 합니다.

그러기 위해서는 데이터를 훈련데이터와 테스트 데이터로 나누어야 하는데...

scikit-learn 에서는 데이터를 섞어서 나누어 주는 train_test_split()를 제공하여 줌
==> 75 대 25로 분리함

그리고 scikit-learn 에서는 데이터는 X(데이터는 2차원 배열(행렬)이므로 대문자로),
타겟은 y(1차원 배열(벡터)이므로 소문자로)를 사용함

```python
# train_test_split 반환값 명명 규칙
# 유사 난수 생성기를 활용하여 데이터셋을 무작위로 섞어야 한다 random_state를 고정함

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], 
                                                    random_state=0)
```

1.7.4 첫번째 머신러닝 모델 : k-최근접 이웃 알고리즘

k는 가장 가까운 이웃 '하나'가 아니라 훈련데이터에서 새로운데이터 포인트에 가장 가까운 'k개'의 이웃을 찾는것

Sci-kit learn의 모든 머신러닝 모델은 Estimator라는 파이선 클래스로 각각 구현 되어 있음

모델을 사용하려면 클래스로 부터 객체를 만들어야 하고 이때 모델에 필요한 매개 변수를 넣는다.

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)

# knn 객체는 훈련데이터로 모델을 만들고 새로운 데이터 포인트에 대해 예측하는 알고리즘을 캡슐화 한 것이다.
# 또한 알고리즘이 훈련 데이터로부터 추출한 정보를 담고 있다.
# KNeighborsClassifier 의 경우는 훈련데이터 자체를 저장하고 있습니다.
```

fit()로 모델 만들기

훈련데이터셋으로 부터 모델을 만들려면 knn 객체에 fit 메서드를 상용한다

```python
knn.fit(X_train, y_train)
```

1.7.5 예측하기

```python
# 길이가 5cm, 폭이 2.9cm, 꽃잎의 길이가 1cm, 폭이 0.2cm인 붓꽃

X_new = np.array([[5, 2.9, 1, 0.2]])
X_new.shape
```

1.7.6 모델 평가하기

앞서 만든 테스트 세트를 사용하여 모델의 정확도를 평가해 보자

```python
y_pred = knn.predict(X_test)
y_pred
```

```python
np.mean(y_pred == y_test)
```

```python
# knn 객체의 score()를 활용하여 정확도를 계산할 수 있다

knn.score(X_test, y_test)
```

1.8 요약 및 정리

```python
# 훈련과 평가가 모두 들어 있는 전체 코드

X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

print("테스트 세트의 정확도: {:.2f}".format(knn.score(X_test, y_test)))
```
